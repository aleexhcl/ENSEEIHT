## Interprétation 

1. Quelle relation lie la fonction test $f_1$ et son modèle de Taylor à l’ordre 2 ? Comparer alors les performances de Newton et RC-Pas de Cauchy sur cette fonction.

2. Le rayon initial de la région de confiance est un paramètre important dans l’analyse
de la performance de l’algorithme. Sur quel(s) autre(s) paramètre(s) peut-on jouer
pour essayer d’améliorer cette performance ? Étudier l’influence d’au moins deux de
ces paramètres.

## Réponses

1. La série de Taylor à l'ordre 2 de la fonction $f_1$ nous donne pour approximation de $f_1$ pour $s$ petit, pour $x \in R^n$, $f(x+s) = f(x) + (\nabla(f)(x))^Ts + (1/2)s^T\nabla^2(f)(x)s$. Le pas de Cauchy permet donc de considérer en plus du gradient les autres directions autour de $x$ tout en restant proche de la valeur de $x$, car l'on reste à une distance $\delta_k$ de $x$ dans cette méthode. Les performances de l'algorithme de Newton sont moins efficaces, car si le gradient coupe l'axe des abcisses en un point trop éloigné de la solution du problème de minimisation, l'algorithme de Newton diverge (comme pour $f_2$ pour certains points initiaux). Le pas de Cauchy permet de rester aux alentours de la solution du problème en donnant une évolution de $x_k$ plus fine.
<br/>
<br/>
2. L'évolution de $\delta_k$ dépend des valeurs de $\gamma_1$ et $\gamma_2$, qui permettent de réduire ou d'augmenter la région de confiance selon l'approximation de $f$ par sa série de Taylor, et la perfomance dépend aussi de $\eta_1$ et $\eta_2$ qui permettent de déterminer si l'approximation de $f$ par sa série de Taylor est assez bonne ou non. 
    - En modifiant le paramètre $\eta_1$ on observe qu'avec un $\eta_1$ plus élevé, la solution trouvée se rapproche davantage de la solution exacte du problème pour un même nombre d'itération. Pour $f_2$ au point initial $x_{023}$, on trouve une solution de $[0.9999370949654659, 0.9998739880304974]$ avec $\eta_1 = 0.1$, et une solution de $[0.9999993488201874, 0.9999986988623017]$ avec $\eta_1 = 0.5$, ce qui rapproche $f_2$ de $0$ à la solution trouvée de $10^{-4}$ environ. L'algorithme converge donc plus rapidement. 
    - En modifiant le pramètre $\gamma_2$ on peut observer qu'en augmentant sa valeur, la solution trouvée est moins proche de la solution exact pour un même nombre d'itération. En effet pour $f_2$ au point initial $x_{023}$, pour $\gamma_2 = 2$ on obtient une solution de $[0.9999370949654659, 0.9998739880304974]$, et avec $\gamma_2 = 5$, on obtient pour solution $[0.9987471516481539, 0.9974937305838003]$, ce qui éloigne $f_2$ de $0$ à la solution trouvée de $10^{-3}$. L'algorithme converge donc moins rapidement.

3. Un problème avec des contraintes d'égalilté et d'inégalité peut s'écrire :
<br/>
    $(P) :$ $min_{x \in R^n}$ $f(x)$ <br/> $c_e(x) = 0$ <br/> $c_i(x) \le 0 $
    <br/> <br/>
    On peut donc écrire les contraintes d'inégalité comme : <br/>
    $c_i(x) + s = 0$ , $s \ge 0$ <br/>
